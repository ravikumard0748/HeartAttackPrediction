```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../datas/newdf.csv")
```

```python
df.head()
```

```python
y=df.pop('Heart Attack Risk')
```

```python
df.head()
y
```

```python
df.pop('Unnamed: 0')
```

```python
from sklearn.utils.class_weight import compute_class_weight
classwe=compute_class_weight(class_weight='balanced',y=y,classes=np.unique(y))
```

```python
classdic={0:classwe[0],1:classwe[1]*}
classdic
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(class_weight=classdic)
```

```python
LR.fit(x_train,y_train)
```

```python
pred=LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred)
```

```python
acc
```

```python
from sklearn.metrics import classification_report,confusion_matrix
clsrep=classification_report(y_test,pred)
conf_mat=confusion_matrix(y_test,pred)
```

```python
print(conf_mat)
```
